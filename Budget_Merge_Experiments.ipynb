```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load nulls final dataframe from previous notebook where we converted all 0s and empty lists to null values
df = pd.read_csv('data/cleaned_csv/nulls_final_merged_df.csv', index_col=0)
```

```python
# Load supplemental budget data
mojo_budget_df = pd.read_csv('data/Mojo_Budget_Data/Mojo_budget_update.csv')
```

```python
mojo_budget_df.head()
```

```python
# Rename movies_df id
mojo_budget_df.rename(columns={'movie_id': 'imdb_id'}, inplace=True)
```

```python
mojo_budget_df.info()
```

```python
# Check how many nulls we have in budget
df['budget'].isna().sum()
```

```python
# Map the 'imdb_id' of df to 'imdb_id' of mojo_budget_df to get the corresponding 'budget'
budget_mapping = df['imdb_id'].map(mojo_budget_df.set_index('imdb_id')['budget'])

# Now, fill the NaN or missing values in df['revenue'] with the values from the mapped 'revenue'
df['budget'].fillna(budget_mapping, inplace=True)
```

```python
# Check again how many nulls we have in budget
df['budget'].isna().sum()
```

Great, we filled in ~200 missing budgets. Not great, but it's something!
